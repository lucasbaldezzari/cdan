```python
data = utils.transform_and_get_iris().drop(columns="segmento")
data.head()
```

```python
data.describe()
```

```python
##entreno Kmeans para 3 clusters
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=42)

kmeans.fit(data) ##entrenamos el modelo
labels = kmeans.labels_ ##etiquetas obtenidas por el modelo
centros = kmeans.cluster_centers_ ##centros de los clusters

intro_colab.plotExample1(data, labels,centros,(8,5))
```
