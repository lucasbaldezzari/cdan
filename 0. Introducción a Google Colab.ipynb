#### **<font color="#d6b302">Celdas</font>**

Una *celda* es simplemente un recuadro donde podemos colocar texto o código.

#### **Celda de texto**

Una celda de texto, como su nombre lo indica, nos permite colocar código (cómo este que estas leyendo).

¿Es posible modificar el contenido de una celda de texto? -> ¡Sí! Para hacerlo, simplemente tenes que hacer doble click y podrás editar, agregar y/o borrar lo que quieras.

#### **Celda de *código***

Este tipo de celdas nos permite colocar código (en lenguaje *python*) y ejecutar el mismo.

*¿Cómo nos damos cuenta que es una celda de código?*

En primer lugar, el color de fondo de estas celdas es ligeramente diferente a las celdas de texto. En segundo lugar, es altamente probable que te des cuenta que se trata de código porque verás cosas como la siguiente:

## <font color="#00b351">Ejemplo de uso de una colab</font>

A continuación, veamos un ejemplo sencilla para entender cómo ejecutar una celda de código.

En este caso, la salida final mostrará dos gráficos de dispersión, los cuales muestran el antes y el después de agrupar un set de datos que utlizaremos de ejemplo en varios momentos del cursado.

````python
## **** CÓDIGO PYTHON ****

##Clonamos el repositorio para poder usar las funciones
!git clone https://github.com/lucasbaldezzari/cdan.git
````

````python
from cdan.funciones import utils ##importo funciones a usar
from cdan.funciones import intro_colab ##importo graficos a usar

data = utils.transform_and_get_iris().drop(columns="segmento")
##entreno Kmeans para 3 clusters
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=42)

kmeans.fit(data) ##entrenamos el modelo
labels = kmeans.labels_ ##etiquetas obtenidas por el modelo
centros = kmeans.cluster_centers_ ##centros de los clusters

intro_colab.plotExample1(data, labels,centros,(8,5))
````

#### **<font color="#d6b302">Resultado</font>**

Como podemos ver, luego de ejecutar las dos celdas de código anterior hemos generado una figura con dos gráficos. El gráfico de la izquierda muestra la distribución de datos (puntos) del set de datos. El gráfico de la derecha, por otro lado, representa el agrupamiento realizado con K-means. Podemos ver que el algoritmo ha generado tres grupos.

Aprenderemos más acerca de cómo se han agrupado los puntos originales en la clase *Algoritmos básicos para clustering* 👌.
